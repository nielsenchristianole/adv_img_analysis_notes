```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

from utils import predict_afine_transform
```

```python
num_points = 10
img_size = (224, 224)

_xs = np.random.randint(0, img_size[0], num_points)
_ys = np.random.randint(0, img_size[1], num_points)
points = np.array([_xs, _ys]).T
```

```python
theta = np.pi
translate = (170, 170)
scale = 0.5

rot_matrix = lambda theta: np.array([[np.cos(theta), -np.sin(theta)],
                                     [np.sin(theta),  np.cos(theta)]])



qoints = scale * (rot_matrix(theta) @ points.T).T + np.array(translate)

plt.plot(*points.T, '.')
plt.plot(*qoints.T, '.')
plt.show()
```
